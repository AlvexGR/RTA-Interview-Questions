```python
df = pd.read_csv('survey_data.csv')
df.head()
```

**Number of interviews each enumerator has completed**

```python
df['nguoipv'].value_counts()
```

```python
df['submission_date'] = pd.to_datetime(df['submission_date'])
df.head()
```

```python
df['starttime'] = pd.to_datetime(df['starttime'])
df['endtime'] = pd.to_datetime(df['endtime'])
```

```python
df['duration'] = df[['starttime', 'endtime']].apply(lambda x: pd.Timedelta(x['endtime'] - x['starttime']), axis = 1)
```

```python
df['duration'] = pd.to_datetime(df['duration'])
```

```python
df['duration'] = df['duration'].apply(lambda x: x.hour * 60 + x.minute + x.second / 60)
```
