**Convert Ngày sinh to a more readable form**

```python
df['Ngày sinh'] = pd.to_datetime(df['Ngày sinh'].apply(lambda x: x.split(' ')[0]))
```

```python
df['Ngày sinh'] = df['Ngày sinh'].apply(lambda x: str(x.day)+ '-' + str(x.month) + '-' + str(x.year))
```

**Acording to Họ tên Ứng viên, we can tell that 1 is male and 2 is female. Now we need to convert column 'Giới tính'**

```python
def convertGender(gender):
    if gender == 1:
        return 'Nam'
    return 'Nữ'
```

```python
df['Giới tính'] = df['Giới tính'].apply(convertGender)
```
